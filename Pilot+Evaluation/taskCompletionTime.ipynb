```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import scienceplots

plt.style.use(['science', 'grid', 'no-latex'])
completion_times = [
    [20,29,23],
    [27,21,20],
    [20,29,23],
    [39,17,34],
    [25,22,21],
    [13,19,12],
    [28,20,22],
    [23,25,20],
    [25,27,23],
    [36,22,23],
]

completion_times = np.array(completion_times)


# Participant labels
participants = [f'P{i+1}' for i in range(10)]
# Assign colors for each participant
# Use the "Blues" colormap with different saturation levels
blues_cmap = cm.get_cmap("Blues", 12)  # 10 different shades of blue
colors = [blues_cmap((i+2) / 10 + 0.2) for i in range(10)]  # Adjust lightness by scaling

plt.figure(figsize=(5, 3))

# Scatter Plot
for i in range(10):  # Iterate through participants
    plt.scatter([participants[i]] * 3, completion_times[i, :], color=colors[i], alpha=0.5, s=50)

# Compute and plot mean completion time
mean_completion_time = np.mean(completion_times, axis=1)
plt.plot(participants, mean_completion_time, marker='^', linestyle='-', color='orange', label='Mean Completion Time', linewidth=2, markersize=10)

# Labels and title
plt.xlabel('Participants')
plt.ylabel('Completion Time (Seconds)')
plt.title('Task Completion Time for Each Participant')
plt.legend()
plt.grid(axis='x')
plt.tight_layout()
plt.savefig('timeCompletion.png', dpi=500)
plt.show()
```

```python
easy_to_use = np.array([5, 5, 5, 4, 5, 4, 4, 4, 5, 5])
recommendation = np.array([5, 5, 5, 4, 5, 5, 5, 5, 5, 5, 5])
satisfaction = np.array([5, 3, 5, 4, 4, 5, 4, 5, 4, 4])

# Compute mean and standard deviation
categories = ["Usability", "Recommendation Likelihood", "Satisfaction"]
means = [easy_to_use.mean(), recommendation.mean(), satisfaction.mean()]
std_devs = [easy_to_use.std(ddof=1), recommendation.std(ddof=1), satisfaction.std(ddof=1)]  # ddof=1 for sample std

# Plotting
plt.figure(figsize=(5, 3))
plt.bar(categories, means, yerr=std_devs, capsize=5, color=["#4C72B0", "#55A868", "#C44E52"], alpha=0.8)

# Labels and title
plt.ylabel("Average Score")
plt.title("Survey Scores with Standard Deviation")
plt.ylim(0, 6)  # Keeping y-axis readable
# plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.grid(False)
plt.tight_layout()
plt.xticks(rotation=15)  # Rotate labels 45 degrees, align right
plt.savefig('surveyScores.png', dpi=500)

plt.show()
```
